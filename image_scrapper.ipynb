```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import numpy as np
```

```python
link="https://stock-pictures.netlify.app"
browser=webdriver.Edge()
browser.get(link)
```

```python
soup=BeautifulSoup(browser.page_source,"html.parser")
```

```python
data=[]
for sp in soup.find_all("div",class_="container"):
    if("gif" not in sp.find("img",class_="source-img").get('src')):
        try:
           img_link=sp.find("img",class_="source-img").get('src')
        except:
           img_link=np.nan
    try:
        Tags=sp.find("div",class_="tags").text
    except:
        Tags=np.nan
    try:
        likes=sp.find("div",class_="likes-comments").find_all("span")[0].text.replace("Likes","")
    except:
        likes=np.nan
    try:
        comments=sp.find("div",class_="likes-comments").find_all("span")[1].text.replace("Comments","")
    except:
        comments=sp.nan
    data.append([Tags,likes,comments,img_link])
        
```

```python
print(data)
```

```python
df=pd.DataFrame(data,columns=["Tags","Likes","Comments","Image_Link"])
```
